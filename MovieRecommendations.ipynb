### 🔎 NLP Preprocessing (What & Why)

Hum movie text data ko clean aur normalize kar rahe hain taaki model ko sirf meaningful words mile.  
Is process me lowercase conversion, punctuation removal, stopwords removal aur lemmatization kiya ja raha hai.  
Isse noise kam hota hai aur similarity calculation zyada accurate ho jaati hai.

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

`re` Python ka Regular Expression library hai jo text me pattern match aur unwanted characters remove karne ke liye use hota hai.

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
stop_words.discard('not')
```

```python
#text preprocessing
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #removed the punctuations
  text = re.sub(r'[^a-zA-Z\s]', '', text)

  words = text.split()
  #remove stopwords
  words = [word for word in words if word not in stop_words]

  #lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python

df['tags'] = df['tags'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

### TF-IDF

TF-IDF text ko numerical vectors me convert karta hai jahan important (rare) words ko zyada weight milta hai aur common words ko kam.  
Isse cosine similarity se better movie recommendations milti hain.

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

### TF-IDF Vectorizer Setup

Hum `TfidfVectorizer` use kar rahe hain taaki cleaned movie text ko meaningful numerical vectors me convert kar sakein.  
`max_features=50000` important words limit karta hai, `ngram_range=(1,2)` single aur two-word phrases capture karta hai, aur `stop_words='english'` common words remove karta hai.

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```


## Cosine Similarity (Intuition + Math)

Cosine similarity do vectors ke **angle** ko measure karta hai.  
Angle jitna **chhota**, similarity utni **zyada**.

\[
\text{Cosine Similarity} = \frac{A \cdot B}{||A|| \, ||B||}
\]

- **cos(0°) = 1 → perfectly similar**
- **cos(45°) ≈ 0.71 → moderately similar**
- **cos(90°) = 0 → completely different**

---

## Example (Movie Text)

### Case 1
**S1:** `virat kohli`  
**S2:** `virat`  

Common word: `virat`

Angle ≈ **45°**



Similarity ≈ **cos(45°) = 0.71**

➡ Meaning: **partially similar**

---

### Case 2
**S1:** `virat`  
**S2:** `kohli`

Common words: **none**

Angle ≈ **90°**



Similarity = **cos(90°) = 0**

➡ Meaning: **not similar**

---

### Case 3
**S1:** `virat kohli`  
**S2:** `virat kohli`

Angle = **0°**



Similarity = **cos(0°) = 1**

➡ Meaning: **exactly same**

---

## Why Important in Your Movie Recommender

TF-IDF movie text ko **vectors** banata hai, aur **cosine similarity** un vectors ke angle compare karke **similar movies recommend karta hai**.

```python
def recommend(title, n = 10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()# flatten sesab 1d bangea
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  #argsort se index basis pe sort hua and index ke basis pe top 10 similar ko choose kiya
  return df['title'].iloc[similar_idx]
```

### Cosine Similarity Formula

\[
\text{Cosine Similarity} =
\frac{\sum_{i=1}^{n} A_i B_i}
{\sqrt{\sum_{i=1}^{n} A_i^2} \times \sqrt{\sum_{i=1}^{n} B_i^2}}
\]

**Variables:**

- \(A_i\) → vector A ka ith element (sentence 1)  
- \(B_i\) → vector B ka ith element (sentence 2)  
- \(n\) → total features / words  

### Example

S1: **virat kohli**  
S2: **virat**

Vectors (simple example):



Angle ≈ **45°**

\[
cos(45^\circ) \approx 0.71
\]

➡ Matlab sentences **partially similar** hain kyuki dono me **“virat”** common hai.
